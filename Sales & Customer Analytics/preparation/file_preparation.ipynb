```python
import pandas as pd
```

```python
customers = pd.read_csv('../dataset/dim_customers.csv')
customers.name = 'Customers'

products = pd.read_csv('../dataset/dim_products.csv')
products.name = 'Products'

sales = pd.read_csv('../dataset/fact_sales.csv')
sales.name = 'Sales'
```

```python
customers.head()
products.head()
sales.head()
```

```python
# Verify column properties

for df in [customers, products, sales]:
    # data types
    print(f"DATABASE: {df.name}")
    print(f'{df.name} data types: {df.info()}')

# primary key
    print(f'''
          -------------------------------
          PRIMARY KEY CHECK
          -------------------------------
          ''')
    for col in df.columns:
        if any(k in col for k in ['id', 'key', 'number']):
            print(f'''
                Duplicates for {col}: {sum(df[col].duplicated())}
                Null count for {col}: {sum(df[col].isnull())}
                ''')
        
    # string length
    print(f'''
          -------------------------------
          STRING LENGHT
          -------------------------------
          ''')
    for col in df.select_dtypes('object').columns:
        if 'date' not in col:
            print(f'''
                {col} max lenght: {df[col].str.len().max()}
                ''')

    # convert dates
    for col in df.columns:
        if 'date' in col:
            df[col] = pd.to_datetime(df[col])
            df[col].dt.strftime('%Y-%m-%d')

    print(f'{df.name} FINAL data types:\n')
    print(df.info())

    print("-"*50)
        
```
